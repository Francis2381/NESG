# manufacturing analysis

```python
manufacture = pd.read_excel('manufacturing.xlsx')
```

```python
manufacture.head()
```

```python
manufacture.info()
```

```python
manufacture['Total government debt'].fillna(manufacture['Total government debt'].mean(), inplace= True)
```

```python
for k in manufacture.columns:
    if manufacture[k].count() != 38:
        manufacture[k].fillna(manufacture[k].mean(), inplace = True)
    else:
        pass
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax= MinMaxScaler()
scaled = minmax.fit_transform(manufacture)
```

```python
to_frame = pd.DataFrame(scaled, columns= manufacture.columns)

y = to_frame['Total Real GDP']
x = to_frame.drop('Total Real GDP', axis = 1)
```

```python
x.head()
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x, y)
lr.score(x, y)
```
